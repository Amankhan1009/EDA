```python
# 1MB = 1024 kB

def size_process(item):
    if str(item)[-1] == "M":
        res = float(str(item).replace("M", ""))
        res = res*1024
        return res
    elif str('item') == "k":
        res = float(str(item).replace("k", ""))
        return res
    else:
        return str(np.NaN)
        
```

```python
size_process('19M')
```

```python
df_copy['Size'] = df_copy['Size'].apply(size_process)
```
